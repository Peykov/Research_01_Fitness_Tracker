### Turn into function

```python
data_path = "../data/raw/MetaMotion\\"
files = glob("../data/raw/MetaMotion/*.csv")


def read_datea_from_files(files):
    acc_df = pd.DataFrame()
    gyr_df = pd.DataFrame()

    acc_set = 1
    gyr_set = 1

    for f in files:
        participant = f.split("-")[0].replace(data_path, "")
        label = f.split("-")[1]
        category = f.split("-")[2]
        category = re.sub(r"\d+", "", category).split("_")[0]

        df = pd.read_csv(f)
        df[["participant", "label", "category"]] = participant, label, category

        if "Accelerometer" in f:
            df["set"] = acc_set
            acc_df = acc_df.append(df)
            acc_set += 1
        elif "Gyroscope" in f:
            df["set"] = gyr_set
            gyr_df = gyr_df.append(df)
            gyr_set += 1

    acc_df.index = pd.to_datetime(acc_df["epoch (ms)"], unit="ms")
    gyr_df.index = pd.to_datetime(gyr_df["epoch (ms)"], unit="ms")

    acc_df.drop(["epoch (ms)", "time (01:00)", "elapsed (s)"], axis=1, inplace=True)
    gyr_df.drop(["epoch (ms)", "time (01:00)", "elapsed (s)"], axis=1, inplace=True)
    return acc_df, gyr_df
```

```python
acc_df, gyr_df = read_datea_from_files(files)
```

```python
gyr_df
```

### Merging datasets (taking first 3 columns of acc_df)

```python
data_merged = pd.concat([acc_df.iloc[:, :3], gyr_df], axis=1)
```

### Resample data (frequency conversion) 
1. renaming the columns
2. We have a lot NaN values because of different sampling rates
- Resample to 200ms (5Hz) and drop the nan values
    - resampling needs an aggregation function (mean for example)
- .apply() sampling rule for every column ( mean doesn't work for categorical)
3. split by chunks of days 
- Accelorometer data is measured 1/12.5 seconds (0.08) 12.5Hz 
Gyroscope data is measured 1/25 seconds (0.04) 25Hz
but they started at different times


```python
data_merged.columns = [
    "acc_x",
    "acc_y",
    "acc_z",
    "gyr_x",
    "gyr_y",
    "gyr_z",
    "participant",
    "label",
    "category",
    "set",
]
```

```python
sampling = {
    "acc_x": "mean",
    "acc_y": "mean",
    "acc_z": "mean",
    "gyr_x": "mean",
    "gyr_y": "mean",
    "gyr_z": "mean",
    "label": "last",
    "category": "last",
    "participant": "last",
    "set": "last",
}
```

```python
data_merged[:100].resample(rule="100ms").apply(sampling)
```

### pd.Grouper (groupby time) most comonly used with resample

- A Grouper allows the user to specify a groupby instruction for an object.
- Aggregate function if we want to have some values of the group count

```python
days = [g for n, g in data_merged.groupby(pd.Grouper(freq="D"))]
days
```

```python
data_resampled = pd.concat(
    [df.resample(rule="200ms").apply(sampling).dropna() for df in days]
)
```
